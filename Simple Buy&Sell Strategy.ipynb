```python
import pandas_datareader as pdr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
start_date = '2014-01-01'
end_date = '2018-01-01'
stock_data = pdr.DataReader('GOOG','yahoo',start_date,end_date)
```

```python
stock_data
```

```python
stock_data_signal = pd.DataFrame(index=stock_data.index)
stock_data_signal
```

```python
stock_data_signal['Price'] = stock_data['Close']
stock_data_signal['Difference'] = stock_data_signal['Price'].diff()
stock_data_signal
```

```python
stock_data_signal['signal'] = np.where(stock_data_signal['Difference']> 0, 1.0, 0.0)
stock_data_signal
```

```python
stock_data_signal['positions'] = stock_data_signal['signal'].diff()
stock_data_signal
```

```python
fig = plt.figure(figsize=(12, 10), dpi=80)
ax1 = fig.add_subplot(111, ylabel='stock Price in $')
stock_data_signal['Price'].plot(ax=ax1, color='r', lw=2.)

ax1

ax1.plot(stock_data_signal.loc[stock_data_signal.positions ==1.0].index,
stock_data_signal.Price[stock_data_signal.positions == 1.0], '^', markersize=5, color='m')

ax1.plot(stock_data_signal.loc[stock_data_signal.positions ==-1.0].index,
stock_data_signal.Price[stock_data_signal.positions == -1.0], 'v', markersize=5, color='k')
```

```python
initial_capital = float(1000.0)
positions = pd.DataFrame(index=stock_data_signal.index)
portfolio = pd.DataFrame(index=stock_data_signal.index)
```

```python
positions['stock'] = stock_data_signal['signal']
positions
```

```python
#axis 0 is vertical, axis = 1 is horizontal
portfolio['positions'] = (positions.multiply(stock_data_signal['Price'], axis =0))
portfolio
```

```python
portfolio['cash'] = initial_capital - (positions.diff().multiply(stock_data_signal['Price'], axis=0)).cumsum()
portfolio['total'] = portfolio['positions'] + portfolio['cash']
portfolio
```

```python
portfolio["total"].plot()
```

```python
fig = plt.figure(figsize=(12, 10), dpi=80)
ax1 = fig.add_subplot(111, ylabel='Portfolio value in $')
portfolio['total'].plot(ax=ax1, lw=2.)
ax1.plot(portfolio.loc[stock_data_signal.positions ==1.0].index,
portfolio.total[stock_data_signal.positions == 1.0], '^', markersize=5, color='m')

ax1.plot(portfolio.loc[stock_data_signal.positions ==-1.0].index,
portfolio.total[stock_data_signal.positions == -1.0], 'v', markersize=5, color='k')
```
